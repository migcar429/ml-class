```python
import matplotlib.pyplot as plt
from scipy.io import loadmat
import numpy as np
from sklearn.svm import SVC


% matplotlib inline
```

```python
data = loadmat('ex6data1.mat')

X, y = data['X'], data['y']

# Tested using a logical index in normal lists vs np arrays (doesn't work for lists).
'''hello = np.array([
    [10,1],
    [32,0]
])

print hello[(hello[:,1:]==0)]'''
```

<b> Plotting to see the effect of C </b>

```python
# Plots a scatter for the given ax. Uses logical indices to grab y==0,y==1 values.
def scatterpl(X, y, ax=plt):
    markers = ['x', '+']
    colors = ['b','r']

    for i, v in enumerate(markers):
        ax.scatter(X[(y==i).ravel(),0], X[(y==i).ravel(),1], marker=v, c=colors[i])
    return None

# Setting up for 
x_mesh = np.linspace(0,5,50)
y_mesh = np.linspace(0,5,50)
xx, yy = np.meshgrid(x_mesh, y_mesh)

fig, ax = plt.subplots(2, figsize=(6,7))

# Default kernel is 'rbf' or radial basis function kernel which is the Gaussian kernel.
# When you instantiate SVC() you can change parameters C and kernel.
for i, v in enumerate([1,100]):
    clf = SVC(C=v, kernel='linear')
    clf.fit(X, y.ravel())
    Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    
    scatterpl(X, y, ax[i])
    ax[i].contourf(xx, yy, Z.reshape(xx.shape), alpha=0.2)
    ax[i].set_title('C={0}'.format(v))
```

<b> Gaussian Kernel </b>

```python
# Not going to use it, but implemented the Gaussian kernel.
from __future__ import division

def gaussian_kernel(x1, x2, sigma):
    
    distance = (x1-x2).T.dot(x1-x2)
    
    g_kerneled = np.exp(-distance/(2*sigma**2))
    
    return g_kerneled
```

<b> Plotting with Gaussian Kernel </b>

```python
data2 = loadmat('ex6data2.mat')
data3 = loadmat('ex6data3.mat')

X2, y2, = data2['X'], data2['y']
X3, y3 = data3['X'], data3['y']

fig, ax = plt.subplots(2, figsize=(10,10))

for i, v in enumerate([(X2, y2),(X3, y3)]):
    
    if i==0:
        x_mesh = np.linspace(0,1.2,100)
        y_mesh = np.linspace(0,1.2,100)
    else:
        x_mesh = np.linspace(-1,.7,100)
        y_mesh = np.linspace(-1,.7,100)
        
    xx, yy = np.meshgrid(x_mesh, y_mesh)
    
    clf = SVC(C=50, gamma=10)
    clf.fit(v[0], v[1].ravel())
    Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    
    scatterpl(v[0], v[1], ax[i])
    ax[i].contourf(xx, yy, Z.reshape(xx.shape), alpha=0.2)
```
