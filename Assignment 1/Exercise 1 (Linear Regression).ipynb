Loading and Plotting Data

```python
# Load as ndarray
data = np.loadtxt('ex1data1.txt', delimiter=',')

# Vectorize
X = np.hstack((np.ones((data.shape[0],1)), data[:,0:1]))
y = data[:,1:]

# Plot in red with no outlines
plt.scatter(X[:,1:], y, c='g', edgecolors='none')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.plot()
```

<b>Modeling the Hypothesis and Cost Functions</b>

```python
def h(x, theta = np.zeros(2).reshape(2,1)):
    x = np.dot(x,theta)
    return x

def compute_cost(x, y, theta = np.zeros(2).reshape(2,1)):
    # size property is number of elements in array. X will not work.
    m = y.size
    
    # Calculate h(x) vector.
    H = h(X,theta.reshape(2,1))
    
    # Weird bug with putting 1/(2*m) as the leading term so divided everything by 2m instead.
    J = np.sum(np.square(H-y)) / (2*m)
    return J

compute_cost(X,y)
```

<b>Performing Gradient Descent </b>

```python
def gradient_descent(X, y, theta = np.zeros(2).reshape(2, 1), iterations = 1500, alpha = 0.01):
    m = y.size
    cost_iterate = []
    learning_term = alpha/m
        
    for iteration in range(iterations):
        # Call hypothesis function with new reshaped theta.
        H = h(X,theta.reshape(2,1))
        
        #theta_j = theta_j - alpha*(1/m) * sum(h(x)-y) * x_j
        theta = theta - learning_term * np.dot(X.T,(H-y))

        # Calculate and store cost for each iteration
        cost_iterate.append(compute_cost(X, y, theta))

    return theta, cost_iterate

theta, cost_iterate = gradient_descent(X,y)
```

```python
# Using a stepped vector for X to plot the regression line.
x_learned = np.arange(1,30)
y_learned = theta[0] + theta[1] * x_learned

plt.scatter(X[:,1:], y, c='g', edgecolors='none')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.plot(x_learned,y_learned, c='r')
```

Help from http://eli.thegreenplace.net/2014/meshgrids-and-disambiguating-rows-and-columns-from-cartesian-coordinates/

```python
# Can use arange or linspace --> http://quantumwise.com/forum/index.php?topic=110.0
x_mesh = np.linspace(-10,10,50)
y_mesh = np.linspace(-1,4,50)
theta_1, theta_2 = np.meshgrid(x_mesh, y_mesh)

# Set matrix where cost values to be stored.
cost_mesh = np.zeros((theta_1.shape[0],theta_1.shape[1]))

# Compute cost for each theta value pair and set the returned value to the appropriate position in cost_mesh.
for i in range(theta_1.shape[0]):
    for j in range(theta_1.shape[1]):
        cost_mesh[i,j] = compute_cost(X, y, theta=np.array((theta_1[i,j],theta_2[i,j])))

# Figsize controls size of plot area for entire figure.
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2, projection='3d')

# Plot contours on specified values of cost via arange function. Alternatively, can also use logspace.
ax1.contour(theta_1, theta_2, cost_mesh, np.arange(-100, 1000, 18))
ax1.scatter(theta[0],theta[1], c='y', edgecolors='none')

# cstride and rstride control intervals of which to sample data to create graph. Alpha controls intensity of color.
ax2.plot_surface(theta_1, theta_2, cost_mesh, cstride=1, rstride=1, cmap=plt.cm.rainbow, alpha=0.5)
# Set elevation (along z-axis), other parameter not given is azim which rotates the figure along Z.
ax2.view_init(elev=20)
```
