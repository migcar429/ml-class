<b> Implementing the Sigmoid and Cost Functions </b>

```python
def sigmoid(z):
    return (1 / (1 + np.exp(-z)))

def compute_cost(theta, X, y):
    m = X.shape[0]
    h = sigmoid(X.dot(theta))
    
    J = ( (-y.T.dot(np.log(h))) - (1-y.T).dot(np.log(1-h)) )/m
    
    # Returning infinity when J gets too large (http://bit.ly/1TbhRiE)
    if np.isnan(J[0]):
        return(np.inf)
    
    return J[0]

# Checking with 3x1 zeros vector.
theta = np.zeros(X.shape[1])
c = compute_cost(theta, X, y)
print c
```

```python
# (1/m) * (h-y) * x_j : Note this is similar to computing the gradient for linear regression except that h is computed differently
def compute_gradient(theta, X, y):
    m = X.shape[0]

    # Have to reshape theta because (h-y) outputs a 100x100 matrix with default theta due to its dimensions of (100,)
    h = sigmoid(X.dot(theta.reshape(-1,1)))

    gradient = ( X.T.dot(h-y) )/m
    gradient = gradient.flatten()
    
    return gradient

# Check values
g = compute_gradient(theta, X, y)
print g
```

```python
# In place of fminunc on MATLAB/Octave, using minimize with method=BFGS.
from scipy.optimize import minimize

# Make sure the objective and gradient functions accept arguments in the order of theta, X, y.
# Also make sure that the isnan check is done. 
output = minimize(compute_cost, theta, args=(X,y), method='BFGS', jac=compute_gradient)
```

```python
# Predicting the output for a student with Exam 1 score of 45 and Exam 2 score of 85.

def predict_admission(exam_score1, exam_score2, x=output.x):
    if sigmoid(np.array((1, exam_score1, exam_score2)).dot(x)) >= 0.5:
        return 1
    else:
        return 0

predict_admission(45, 85)
```
